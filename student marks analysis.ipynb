```python
# gender distribution 
```

```python
plt.figure(figsize = (4,3))
sns.countplot(data= df, x = "Gender")
plt.show()
```

```python
plt.figure(figsize = (5,5))
ax= sns.countplot(data= df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

```python
#from the above chart we have analysed  that :
#the number og females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot = True)
plt.title("Relation between parent's Education and  Student's score")
plt.show()
```

```python
# from the above chart we have concluded that:
#the education of the parents have a good impact on scores 
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1,annot = True)
plt.title("Relation between parent's maritalstatus and  Student's score")
plt.show()
```

```python
# from the above chart we have  concluded that  the  parents marital status dosnt have any impact on studies
```

# Distribution  of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
print(groupA)
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l =["group A","group B","group C","group D","group E"]
my_list =[groupA["EthnicGroup"] ,groupB ["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(my_list, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of ethnic groups")
plt.show()
```
